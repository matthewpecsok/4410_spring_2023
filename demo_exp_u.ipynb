```python
import pandas as pd
import sqlite3 
```

```python
con = sqlite3.Connection('northwind.db')
```

```python
pd.read_sql_query("select * from customers",con)
```

```python
pd.read_sql_query("select count(1) as count,Country from customers group by Country",con).sort_values(by='count').plot(kind='barh',x='Country',y='count')
```

```python
pd.read_sql_query("""

select count(1) from customers where Country='USA'

""",con)
```
